# create_co_occurance_matrix

```python
mydict = preprocess.create_paper_paper_dict(debug=True)
```

```python
co_matrix = preprocess.create_co_occurrence_matrix(mydict)
```

```python
len(co_matrix)
```

```python
# Make prediction for a user, using co_occurance_matrix
user_row = list(mydict.values())[0]
user_row_dict = utils.ListToDict(user_row)
prediction = dict()
for key, value in co_matrix.items():
    prediction[key]=utils.dotProduct(value, user_row_dict)
```

```python
# Display created prediction. The bigger the value, the higher scored recommendation.
for key, value in prediction.items():
    if value != 0:
        print(key, value)
```

```python
# Display created prediction, except for the papers this row is already citing
for key in user_row:
    del prediction[key]
sorted_prediction = sorted(prediction.items(), key=operator.itemgetter(1))
for key, value in reversed(sorted_prediction):
    if value != 0:
        print(key, value)
```
